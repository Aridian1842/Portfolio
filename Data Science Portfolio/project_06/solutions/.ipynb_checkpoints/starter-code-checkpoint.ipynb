## Part 3: Build the Decision Tree

#### 1. What is our target attribute? 

Our target attribute is going to be rating.

#### 2. Prepare the data and define the training set

```python
col = df1.columns
```

We wanted to use both attributes and votes in our model, and accomplished this using a pd.concat method.

```python
y = df1['rating']
X1 = df1[col[2:10]]
X2 = df1[col[15]]
X = pd.concat([X1, X2],axis=1)
```

#### 2. Train the Model

Due to the data set being so small - only 25 movies - excluding a subset as the traintest will result in an even smaller number of training points.  Therefore, we decided to use k-fold cross validation in our model.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.cross_validation import KFold, cross_val_score


cv = KFold(len(y), shuffle=True)

dt = DecisionTreeRegressor()

scores = cross_val_score(dt, X, y, cv=cv)
```

```python
#treereg.fit(X_train, y_train)
```

#### 3. Set up test data and test the model

Because we used a k-fold cross validation we can now build a decision tree regressor:

```python
def do_plot(model, m=None):
    for fold, color in zip(cv, ['r', 'g', 'b']):
        X_train = X.iloc[fold[0]]
        X_test  = X.iloc[fold[1]]
        y_train = y.iloc[fold[0]]
        y_test  = y.iloc[fold[1]]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        plt.scatter(y_test, y_pred, color=color)
        plt.plot([8.5,9.5],[8.5,9.5])
        plt.text(9,9.5, "R2: "+str(m), fontsize=20, )
        
do_plot(dt, scores.mean().round(2))
```

#### 5. Check the results

Using a Random Forest Regressor we are able to get tighter results.

```python
from sklearn.ensemble import RandomForestRegressor


rf = RandomForestRegressor()

scores = cross_val_score(rf, X, y, cv=cv)
scores
```

```python
do_plot(rf, scores.mean().round(2))
```

An AdaBoost Regressor refines our results even more:

```python
from sklearn.ensemble import AdaBoostRegressor


ab = AdaBoostRegressor()

scores = cross_val_score(ab, X, y, cv=cv)
scores
```

```python
do_plot(ab, scores.mean().round(2))
```

Out of curiosity I ran a Gradient Boosting Regressor, but that returned worse results:

```python
do_plot(ab, scores.mean().round(2))
```

```python
all_scores = []
best_score = -1
best_depth = 0
for i in range(1, 9):
    treereg = DecisionTreeRegressor(max_depth=i, random_state=1)
    scores = cross_val_score(treereg, X, y, cv=3, scoring='mean_squared_error')
    current_score = np.mean(np.sqrt(-scores))
    # If the score mean is better than the current best, or best is the default (-1), then update!
    if current_score < best_score or best_score == -1:
        best_score = current_score
        best_depth = i
    # store to plot anyway!
    all_scores.append(current_score)
    
print "Best score: %s" % best_score
print "Best depth: %s" % best_depth

# now actually fit the model
treereg = DecisionTreeRegressor(max_depth=best_depth, random_state=25)
treereg.fit(X, y)

plt.figure()
plt.plot(range(1, 9), all_scores)
```
